```python
import numpy as np
import pandas as pd
df = pd.read_csv('..//Data/Unlabelled.csv')
```

```python
df
```

```python
df1 = pd.DataFrame(df,columns=['Operator','Network Type','In Out Travelling','State Name','Call Dropped','Satisfactory','Poor Voice Quality'])
```

```python
df1
```

```python
df1['unsatisfactory']=df1['Call Dropped'] + df1['Poor Voice Quality']
```

```python
df1
```

```python
df1 = pd.DataFrame(df1,columns=['Operator','Network Type','In Out Travelling','unsatisfactory'
,'Satisfactory'])
df1
```

```python
df1.loc[df1['Satisfactory']<=100, 'label'] = 'Excellent'
df1.loc[df1['Satisfactory']<=80, 'label'] = 'Very Good'
df1.loc[df1['Satisfactory']<=60, 'label'] = 'Average'
df1.loc[df1['Satisfactory']<=40, 'label'] = 'Poor'
df1.loc[df1['Satisfactory']<=20, 'label'] = 'Inadequate'
```

```python
df1
```

```python
Label = np.array(df1['label'])
```

```python
df['label']= Label
df
```

```python
df.to_csv('..//Data/Labelled.csv',index=False)
```
