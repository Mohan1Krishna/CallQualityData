```python
import pandas as pd
```

```python
df=pd.read_csv("..//Data/Final_data.csv")
```

```python
df.head(10)
```

operator label is assigned to each operatot(OP_label)

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

NO_le = LabelEncoder()
NO_labels = NO_le.fit_transform(df['Operator'])
df['OP_Label'] = NO_labels
```

```python
df_sub = df[['Operator', 'Network Type', 'OP_Label',  
                       'In Out Travelling', 'Call Drop Category']]
df_sub.iloc[4:15]
```

```python
OP_ohe = OneHotEncoder()
OP_feature_arr = OP_ohe.fit_transform(
                              df[['OP_Label']]).toarray()
OP_feature_labels = list(NO_le.classes_)
OP_features = pd.DataFrame(OP_feature_arr, 
                            columns=OP_feature_labels)
```

```python
OP_df_ohe = pd.concat([df_sub, OP_features], axis=1)
columns = sum([['Operator', 'Network Type', 'OP_Label'],   
               OP_feature_labels], [])
OP_df_ohe[columns].iloc[4:20]
```

OP_label using one-hot encoding are converted into attributes  (nominal are converted into binary attributes)
