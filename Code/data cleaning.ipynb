```python

import pandas as pd
import os
df_list = []

#combing the different datasets. os.listdir will get name of  all files in directory
for file in os.listdir("..//Data"):
    print(file)
    #read_csv imports the data in the file
    df = pd.read_csv("..\Data" +"\\"+ file)
    #attaching data to the list
    df_list.append(df)
#concat will combine data and to_csv will create an output file    
join = pd.concat(df_list)
print(join)
join.to_csv("..//Data/rawdata.csv")
```
