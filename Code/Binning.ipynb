```python
import numpy as np
import pandas as pd
df = pd.read_csv("..//Data/Labelled.csv")
print(df)
```

```python
df['bins'] = pd.cut(df['Satisfactory'],5)
df1 = df.groupby(['State Name','Operator','bins']).size()
df1
```

```python
df1 = df1.reset_index()
df1 = pd.DataFrame(df1)
df1.columns = ['State','operator','bin','size']
df1
```

```python
df2 = df1.groupby(['State','operator'])['size'].sum()
```

```python
df2 = pd.DataFrame(df2)
```

```python
df2.columns = ['size2']
```

```python
df2 = df2.reset_index()
```

```python
res = pd.merge(df1,
                 df2[['State', 'operator','size2']],
                 on=['State','operator'], 
                 how='left')
```

```python
res
```

```python
res['fraction'] = res['size']/res['size2']
```
