```python
import pandas as pd
```

```python
df=pd.read_csv("..//Data/Final_data.csv")
```

```python
df.head()
```

```python
from sklearn.feature_extraction import FeatureHasher
fh = FeatureHasher(n_features=4, input_type='string')
hashed_features = fh.fit_transform(df['Operator'])
hashed_features = hashed_features.toarray()
pd.concat([df[['Operator', 'Network Type','Call Drop Category']], pd.DataFrame(hashed_features)], 
          axis=1).iloc[1:20]
```

Here there are 9 operators they are converted into 4 attributes instead of 9 by feature extraction library (Dimensionality Reduction)
