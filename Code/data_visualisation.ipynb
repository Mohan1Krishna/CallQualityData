Below are pie-charts with % share of call drop category of different networks

```python
import pandas as pd
df = pd.read_csv("..//Data/Final_data.csv")
contingency_table2 = pd.crosstab(
    df['Operator'],
    df['Call Drop Category'],
    margins = True
) # contigency table of frequencies  with operators as rows and call drop category as columns
list = ["airtel","BSNL","Idea","MTNL","Other","RJio","Tata","Telenor","Vodafone"]
print(contingency_table2)
for i in range(8):
    slices = contingency_table2.iloc[i][0:3].values
    operators = ['Call Dropped','Poor Voice Quality','Satisfactory']
    cols = ['c','m','r']
    plt.pie(slices,
               labels=operators,
               colors=cols,
               startangle=90,
               shadow= True,
               autopct='%1.1f%%')
    plt.axis('equal')
    plt.title(list[i])
    plt.savefig("..//Plots/"+list[i]+"callDropQuality.png")
    plt.show()
```
